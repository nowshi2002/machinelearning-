```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Define the dataset name (since it’s in the same directory)
file_name = 'customers.csv'

# Load the dataset
df = pd.read_csv(file_name)

# Display the first few rows of the data
print("First few rows of the dataset:")
print(df.head())

# Preprocess the data by scaling it
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Select the number of clusters
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the model
kmeans.fit(scaled_data)

# Add the cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Plot the clusters (using the first two scaled features for simplicity)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=df['Cluster'], palette='viridis')
plt.title("KMeans Clustering on Wholesale Customers Data")
plt.xlabel("Scaled Feature 1")
plt.ylabel("Scaled Feature 2")
plt.show()
```
